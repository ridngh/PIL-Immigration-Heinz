```python
import pandas as pd
import spacy
from transformers import pipeline

# Load spaCy English model
nlp_spacy = spacy.load("en_core_web_sm")

# Load BERT NER pipeline
nlp_bert_ner = pipeline("ner", model="dbmdz/bert-large-cased-finetuned-conll03-english")
```

```python
# Function to perform spaCy NER
def spacy_ner(text):
    doc = nlp_spacy(text)
    entities = [(ent.text, ent.label_) for ent in doc.ents]
    return entities

def extract_geographical_areas(text):
    doc = nlp_spacy(text)
    return [ent.text for ent in doc.ents if ent.label_ == 'GPE']
```

```python
# Function to extract entity labels using spaCy NER
def extract_entity_labels(text):
    doc = nlp_spacy(text)
    entities = {ent.label_: [] for ent in doc.ents}
    
    for ent in doc.ents:
        entities[ent.label_].append(ent.text)
    
    return entities
```

```python
df = Jobs.sample(300,random_state=1)
```

```python
# Apply the function to the 'jobs_description' column
df['spacy_ner_entities'] = df['description'].apply(extract_entity_labels)

# Create new columns for each entity label
for label in df['spacy_ner_entities'].iloc[0].keys():
    df[label] = df['spacy_ner_entities'].apply(lambda x: ', '.join(x[label]) if label in x else None)

# Drop the temporary 'spacy_ner_entities' column
df = df.drop(columns=['spacy_ner_entities'])
```
