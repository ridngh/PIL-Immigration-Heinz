```python
# Create a function remove HTML tags
from bs4 import BeautifulSoup
def remove_html_tags(text):
    soup = BeautifulSoup(text, 'html.parser')
    return soup.get_text()

```

### Scaling this to scraped job descriptions

```python
def answer_question(row,questions):
    context = row['description']
    answers = []
    for question in questions:
        answer = qa_pipeline(question=question, context=context)['answer']
        answers.append(answer)

    return pd.Series(answers, index=questions)
```

```python
#Try with sample 
questions = ["Can minors be employed here", "Does this provide accomodation", "Does this have educational requirements", "What is the salary range offered in dollars per year format"]

sample[questions] = sample.apply(lambda row: answer_question(row, questions), axis=1)
```

### Named Entity Recogntion on the flagged posts 

```python
# Load the NER pipeline
ner_pipeline = pipeline('ner', model='dbmdz/bert-large-cased-finetuned-conll03-english')

def extract_entities(text):
    entities = ner_pipeline(text)
    entity_dict = {'name': [], 'org': [], 'gpe': []}

    for ent in entities:
        entity_type = ent['entity']
        entity_text = ent['word']

        if entity_type in ('B-PER', 'I-PER'):
            entity_dict['name'].append(entity_text)
        elif entity_type in ('B-ORG', 'I-ORG'):
            entity_dict['org'].append(entity_text)
        elif entity_type in ('B-GPE', 'I-GPE'):
            entity_dict['gpe'].append(entity_text)

    return pd.Series(entity_dict)
```

```python
# Apply the function to extract entities and create new columns
sample[['name', 'org', 'gpe']] = sample['description'].apply(extract_entities)

# Print the resulting DataFrame
sample.head()
```

```python
import re
def extract_websites_and_emails_from_df(data_frame, text_column):
    # Regular expression patterns for websites and email addresses
    website_pattern = re.compile(r'https?://\S+|www\.\S+')
    email_pattern = re.compile(r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b')

    # Initialize empty lists to store results
    website_results = []
    email_results = []

    # Loop through each row in the DataFrame
    for index, row in data_frame.iterrows():
        description_text = row[text_column]

        # Find all matches in the description text
        websites = website_pattern.findall(description_text)
        emails = email_pattern.findall(description_text)

        # Append the results to the lists
        website_results.append(','.join(websites))
        email_results.append(','.join(emails))

    # Add new columns to the DataFrame
    data_frame['website'] = website_results
    data_frame['email'] = email_results

    return data_frame
```

```python
# Example usage:
sample2 = extract_websites_and_emails_from_df(sample, 'description')
# Print the updated DataFrame
sample2.tail()
```

### Creating Location Based Insights from OSIS

```python
import matplotlib.pyplot as plt

def plot_job_counts_by_city(data_frame, target_state):
    # Filter the DataFrame based on the specified state
    state_filtered_df = data_frame[data_frame['location_2'] == target_state]

    # Group by city and count the number of jobs
    city_job_counts = state_filtered_df['location_3'].value_counts()

    # Plotting the bar chart
    plt.figure(figsize=(20, 10))
    city_job_counts.plot(kind='bar', color='skyblue')
    plt.title(f'Job Counts by City in {target_state}')
    plt.xlabel('City')
    plt.ylabel('Job Count')
    plt.xticks(rotation=45, ha='right')
    plt.show()
```

```python
def filter_jobs_by_location(data_frame, target_state):
    # Filter the DataFrame based on the specified state
    state_filtered_df = data_frame[data_frame['location_2'] == target_state]
    return(state_filtered_df)
```

```python
num_splits = 3
new_columns = [f'location_{i+1}' for i in range(num_splits)]
sample[new_columns] = sample['location'].str.split(',', expand=True, n=num_splits-1)
sample[new_columns] = sample[new_columns].apply(lambda x: x.str.strip())
#sample[['Country', 'State']] = sample['location'].str.split(',', expand=True)
sample.head()
```

```python
def generate_col_value_list(data_frame,col_name):
    distinct_values = ','.join(set(item for sublist in data_frame[col_name].dropna().apply(lambda x: x if isinstance(x, list) else [x]) for item in sublist))
    return distinct_values
```
