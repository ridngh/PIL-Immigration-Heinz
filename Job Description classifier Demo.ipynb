```python
# Function to extract entity labels using spaCy NER
def extract_entity_labels(text):
    doc = nlp_spacy(text)
    entities = {ent.label_: [] for ent in doc.ents}
    
    for ent in doc.ents:
        entities[ent.label_].append(ent.text)
    
    return entities
```
